#  1. ¿Qué es la Estadística Inferencial?
#
# La **estadística inferencial** se centra en obtener conclusiones sobre una población a partir de una muestra de datos. Nos permite:
#
# - Estimar parámetros poblacionales (media, proporción, varianza).
# - Realizar pruebas de hipótesis.
# - Construir intervalos de confianza.
# - Hacer predicciones (regresión, series de tiempo).
#
#  En logística, la estadística inferencial es clave para:
# - Predecir la demanda de productos.
# - Optimizar rutas de transporte.
# - Analizar tiempos de entrega.
# - Evaluar costos y eficiencia de procesos.

```python
# %% [code]
# Importamos las librerías básicas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de gráficos
sns.set(style="whitegrid", palette="muted")
```


#  3. Mini ejemplo práctico
#
# Supongamos que queremos analizar la relación entre la **distancia de entrega** (km) 
# y el **costo de transporte** (USD).
#
# Para este ejemplo, generaremos un pequeño conjunto de datos ficticios.

```python

# Creamos un dataset de ejemplo
data = pd.DataFrame({
    "Distancia_km": [5, 10, 15, 20, 25, 30, 35, 40, 45, 50],
    "Costo_USD": [50, 65, 80, 90, 105, 120, 135, 150, 160, 175]
})

data
```

```python

# Visualizamos los datos
plt.figure(figsize=(6,4))
sns.scatterplot(x="Distancia_km", y="Costo_USD", data=data, s=100)
plt.title("Relación entre distancia y costo de transporte")
plt.show()
```


#  Observamos que existe una tendencia lineal creciente: 
# a mayor distancia, mayor costo de transporte.
#
# Más adelante (Módulo 2) formalizaremos este análisis con **regresión lineal**.
